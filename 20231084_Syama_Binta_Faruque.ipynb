<h1>Training</h1>

```python
X=Data[['Open','High','Low','Volume']]
y=Data['Close'] #prediction
```

###Train -Test Split


*   20% of data is splitted as test and 80%data is considered for test

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

###Model loading

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train) #training
```

```python
print('Coefficients (m)',lm.coef_) #printing the coefficients
```

<h1>Prediction</h1>

```python
predictions=lm.predict(X_test)  #testing
```

<h1>Evaluating the Model</h1>
For linear regression, parameters for evaluation is Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
RMSE = float(format(np.sqrt(mean_squared_error(y_test, predictions)),'.3f'))
MSE=mean_squared_error(y_test,predictions)
MAE = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test,predictions)
k=X_test.shape[1]
n=len(X_test)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print("RMSE:",RMSE, "\nMSE:",MSE,"\nMAE:",MAE,"\nr2:",r2,"\nAdjusted r2:",adj_r2)
```

<h1>Ridge Regression</h1>

Finding Alpha using Hyperparamater tuning. Choosing optimal value of alpha using gridsearch

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
folds=KFold(n_splits=5,shuffle=True)
params={'alpha':[0.0001,0.001,0.01,0.1,0.2,0.5,0.9,1.0,5.0,10.0,500.0]}
regressor_ridge=Ridge()
model_cv=GridSearchCV(estimator=regressor_ridge,param_grid=params,scoring='r2',cv=folds, return_train_score=True,verbose=1)
model_cv.fit(X_train,y_train)  #training with ridge regression
```

```python
model_cv.best_params_
```

```python
regressor_ridge=Ridge(alpha=0.0001)
regressor_ridge.fit(X_train,y_train)
print("Coefficient:",regressor_ridge.coef_)
print("Intercept:",regressor_ridge.intercept_)
```

```python
predictions=regressor_ridge.predict(X_test) #testing
```

###Performance metrices for Ridge Regression


```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test, predictions)),'.3f'))
MSE=mean_squared_error(y_test,predictions)
MAE = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test,predictions)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print("RMSE:",RMSE, "\nMSE:",MSE,"\nMAE:",MAE,"\nr2:",r2,"\nAdjusted r2:",adj_r2)
```

```python
#Testing for single data
test_value=[[19,25,17,18766300]]
print(test_value)
predictions=regressor_ridge.predict(test_value)
predictions
```

<h1>Lasso Regression</h2>

Finding Alpha using Hyperparamater tuning. Choosing optimal value of alpha using gridsearch


```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
folds=KFold(n_splits=5,shuffle=True)
params={'alpha':[0.001,0.01,0.1,0.2,0.5,0.9,1.0,5.0,10.0,20]}
regressor_lasso=Lasso()
model_cv=GridSearchCV(estimator=regressor_lasso,param_grid=params,scoring='r2',cv=folds, return_train_score=True,verbose=1)
model_cv.fit(X_train,y_train)
```

```python
model_cv.best_params_
```

```python
regressor_lasso = Lasso(alpha = 0.5)
regressor_lasso.fit(X_train,y_train)
print('Coefficient (m): ', regressor_lasso.coef_)
print('Coefficient (b): ', regressor_lasso.intercept_)

predictions = regressor_lasso.predict( X_test) #testing
```

###Performance metrices for LASSO Regression

```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test, predictions)),'.3f'))
MSE=mean_squared_error(y_test,predictions)
MAE = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test,predictions)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print("RMSE:",RMSE, "\nMSE:",MSE,"\nMAE:",MAE,"\nr2:",r2,"\nAdjusted r2:",adj_r2)
```

```python
test_value=[[19,25,17.540001,18766300]]
print(test_value)
predictions=regressor_lasso.predict(test_value)
predictions
```
