```python
# 10. Perform Exploratory Data Analysis (EDA) using groupby/pivot_table and barplot (total 9 barplots),based on model, transmission, and fuelType,
# a) What are the top 5 selling car models/transmission/fuelType in the dataset?
# b) What's the average selling price of the top 5 selling car models/transmission/fuelType?
# c) What's the total sale of the top 5 selling car models/transmission/fuelType?

import pandas as pd
import matplotlib.pyplot as plt
# Top 5 selling car models
top_5_models = df['model'].value_counts().head(5)

# Top 5 selling transmission types
top_5_transmissions = df['transmission'].value_counts().head(5)

# Top 5 selling fuel types
top_5_fuel_types = df['fuelType'].value_counts().head(5)

# Average selling price of top 5 car models
avg_price_models = df.groupby('model')['price'].mean().loc[top_5_models.index]

# Average selling price of top 5 transmission types
avg_price_transmissions = df.groupby('transmission')['price'].mean().loc[top_5_transmissions.index]

# Average selling price of top 5 fuel types
avg_price_fuel_types = df.groupby('fuelType')['price'].mean().loc[top_5_fuel_types.index]

# Total sale of top 5 car models
total_sale_models = df.groupby('model')['price'].sum().loc[top_5_models.index]

# Total sale of top 5 transmission types
total_sale_transmissions = df.groupby('transmission')['price'].sum().loc[top_5_transmissions.index]

# Total sale of top 5 fuel types
total_sale_fuel_types = df.groupby('fuelType')['price'].sum().loc[top_5_fuel_types.index]

# Plotting barplots
plt.figure(figsize=(12, 8))

# Top 5 selling car models
plt.subplot(3, 3, 1)
top_5_models.plot(kind='bar')
plt.title('Top 5 Selling Car Models')
plt.xlabel('Car Models')
plt.ylabel('Number of Sales')

# Average selling price of top 5 car models
plt.subplot(3, 3, 2)
avg_price_models.plot(kind='bar')
plt.title('Average Selling Price of Top 5 Car Models')
plt.xlabel('Car Models')
plt.ylabel('Average Selling Price')

# Total sale of top 5 car models
plt.subplot(3, 3, 3)
total_sale_models.plot(kind='bar')
plt.title('Total Sale of Top 5 Car Models')
plt.xlabel('Car Models')
plt.ylabel('Total Sale')

# Top 5 selling transmission types
plt.subplot(3, 3, 4)
top_5_transmissions.plot(kind='bar')
plt.title('Top 5 Selling Transmission Types')
plt.xlabel('Transmission Types')
plt.ylabel('Number of Sales')

# Average selling price of top 5 transmission types
plt.subplot(3, 3, 5)
avg_price_transmissions.plot(kind='bar')
plt.title('Average Selling Price of Top 5 Transmission Types')
plt.xlabel('Transmission Types')
plt.ylabel('Average Selling Price')

# Total sale of top 5 transmission types
plt.subplot(3, 3, 6)
total_sale_transmissions.plot(kind='bar')
plt.title('Total Sale of Top 5 Transmission Types')
plt.xlabel('Transmission Types')
plt.ylabel('Total Sale')

# Top 5 selling fuel types
plt.subplot(3, 3, 7)
top_5_fuel_types.plot(kind='bar')
plt.title('Top 5 Selling Fuel Types')
plt.xlabel('Fuel Types')
plt.ylabel('Number of Sales')

# Average selling price of top 5 fuel types
plt.subplot(3, 3, 8)
avg_price_fuel_types.plot(kind='bar')
plt.title('Average Selling Price of Top 5 Fuel Types')
plt.xlabel('Fuel Types')
plt.ylabel('Average Selling Price')

# Total sale of top 5 fuel types
plt.subplot(3, 3, 9)
total_sale_fuel_types.plot(kind='bar')
plt.title('Total Sale of Top 5 Fuel Types')
plt.xlabel('Fuel Types')
plt.ylabel('Total Sale')

plt.tight_layout()
plt.show()
```
